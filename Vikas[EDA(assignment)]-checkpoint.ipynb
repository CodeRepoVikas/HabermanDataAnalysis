```python
sns.FacetGrid(df,hue='status',size=6).map(sns.distplot,'year').add_legend().set(title='PDF of Feature year',xlabel='YEAR', ylabel='Density')
```

<h4>Observation-</h4>

From the Above we can conclude that both the pdfs are overlaping each other and hence we can't categorize status on feature age basis.

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue='status',size=6).map(sns.distplot,'nodes').add_legend().set(title='PDF of feature nodes', xlabel='NODES',ylabel='Density');
```

<strong><h4> Observations- </h4></strong>
<ol><strong>From above PDF</strong>
<li>There are almost 50% of the people having node=0 and lived min 5 years.</li>
<li>A large majority of obsevations have node value less than 5 and lived min 5 years. </li>
<li>Using  if conditions we can say <strong>if(0<=nodes< 5)then "Survived min 5 years" else "didnot survived 5 Years" </strong></li> </ol>   

Performing CDF on each features

```python
status_1=df.loc[df.status==1]
status_2=df.loc[df.status==2]
```

```python
count,bin_=np.histogram(status_1.nodes,density=True,bins=12)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_[1:],pdf,label='PDF')
plt.plot(bin_[1:],cdf,label='CDF')
plt.title("PDF And CDF of feature Nodes When people lived minimum 5 Years")
plt.ylabel("Density")
plt.xlabel("Nodes")
plt.legend(loc='middle left')
plt.show()
print(bin_)
print(count)
```

<strong><h4>Observation-</h4></strong> 
<ol><li>Amlost 80% of the people who lived min 5 years had node value between 0 and 4.</li>
    <li>Almost 100% of the people had node value less than 30 who lived min 5 years.</li>
    <li>almost 98% of the people who lived min 5 years had node value less than 20.</li><ol>

```python
count,bin_=np.histogram(status_2.nodes,density=True,bins=12)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_[1:],pdf,label='PDF')
plt.plot(bin_[1:],cdf,label='CDF')
plt.title("PDF and CDF of feature nodes Where people lived less than 5 years")
plt.xlabel("Density")
plt.ylabel("Nodes")
plt.legend(loc='center right')
plt.show()
print(bin_)
print(count)
```

<strong><h4>Observation-</h4></strong>
<ol><li>Almost 50% of the people had node value between 0 and 4, who didn't survived 5 years.</li>
    <li>85% of the people had node range(0,17) and didn't survived 5 years.</li>
    <li>Almsot 98% of the people had node range(0,26) who didn't survived 5 years. </li>
    <li>99.5% of the people had node value between 0 and 39, who survived less than 5 years.</li></ol> 

```python
count,bin_=np.histogram(status_1.nodes,density=True,bins=12)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_[1:],pdf,label='Pdf(when People survived min 5 years)')
plt.plot(bin_[1:],cdf,label='cdf(when People survived min 5 years)')
print("FOR Surviving more 5 or more years")
print("bins"+str(bin_))
print("count" + str(count))
count,bin_=np.histogram(status_2.nodes,density=True,bins=12)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_[1:],pdf,label='Pdf(when People survived less than 5 years)')
plt.plot(bin_[1:],cdf,label='cdf(when People survived less than 5 years)')
print("\n\nFOR Surviving less than 5 years")
print("bins"+str(bin_))
print("count" + str(count))
plt.title("PDF and CDF of feature nodes ")
plt.xlabel("Nodes")
plt.ylabel("Density")
plt.legend(loc='center right')
plt.show()
```

<strong><h4>Observations-</h4></strong>
<ol><h4>From the above graph and observations we made in the above cdf and pdf graphs we can conclude that-</h4>
    <li>Between node range(0,4) 80% people survived and nearly 50% didnot survived min 5 years.</li>
    <li>People having node value below 20 almost 98% survived and nearly 90% didn't survived min 5 years.</li>
    <li>People having node value in range(0,27) almost 99% people survived and almost 98% didnot survived min 5 years.</li></ol>

## Performing Mathematical Analysis on DataSet

### Mean, Variance and Std.Deciaton

```python
print("mean(where status=1)  "+str(np.mean(df.nodes[df.status==1])));
print("mean(where status=2)  "+str(np.mean(df.nodes[df.status==2])));
print("std.Dev (where status=1)  "+str(np.std(df.nodes[df.status==1])));
print("Std.Dev(where status=2)  "+str(np.std(df.nodes[df.status==2])));
```

<strong><h4>Obervations-</h4></strong> 
<h4>Since we have outliers in our dataset we can see observe that-</h4><ol>
    <li>Mean is 3 and 7 for status=1 and status=2 respectively.</li>
    <li>Std Deviation is 5 and 9 for status= 1 and status=2 respectively.</li></ol>


###  Median, Percentile

```python
print("meadian(where status=1)  "+str(np.median(df.nodes[df.status==1])));
print("meadian(where status=2)  "+str(np.median(df.nodes[df.status==2])));
```

<strong><H4>Observation-</H4></strong>
<ol><li>Median is 0 and 4 for status=1 and status=2 respectively.</li>
   </ol>

```python
print('OBSERVATIONS-')
print("Percentile(where status=1)  "+str(np.percentile(df.nodes[df.status==1],np.arange(0,101,25))));
print("Percentile(where status=2)  "+str(np.percentile(df.nodes[df.status==2],np.arange(0,101,25))));
print("Percentile(where status=1) 95 %  "+str(np.percentile(df.nodes[df.status==1],95)));
print("Percentile(where status=2) 95 %  "+str(np.percentile(df.nodes[df.status==2],95)));
```
